```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline

def pdf(time):
    rate = 3.053*60 
    x = np.linspace(0, time) 
    y = 1-2**(-1*x/rate)
    plt.plot(x,y)
    plt.show()   
    return
def exp_distribution_generator(N):
    hlf = 3.053*60 
    Mean = np.log(2)/hlf
    X = []
    for i in range(N):
        z = random.random()
        x_result = int((-1 / Mean) * np.log(1 - z))
        X.append(x_result)
    plt.hist(X, bins = 100)
    plt.xlim(0, 1000)
    plt.show()
    return
```

```python
pdf(500)
```

```python
exp_distribution_generator(500)
```

```python
def decay(time): # radioactive decay function

    hlf = 3.053*60 
    atom_init = 1000
    Mean = np.log(2)/hlf
    
    t = np.linspace(0, time)
    f = lambda t: atom_init * 2**(-1*t/hlf)
    y = f(t)
    
    plt.plot(t,y)
    plt.show()

# Non uniform

    DECAY = atom_init * np.ones([time - 1])
    PROB = np.zeros([time - 1])
    Nrandompoints = 1000
    
    for i in range(Nrandompoints):
        z = random.random()
        x_result = int((-1 / Mean) * np.log(1 - z))
        
        if x_result < time :
            DECAY[x_result:] -= 1
            PROB[x_result:] += 1
        
    plt.plot(DECAY, label='Decay of Thallium')
    plt.plot(PROB, label='Probability')
    plt.legend()
    plt.show()
    
    return
```

```python
decay(500)
```

## Brownian Motion

```python
L1 = 101
L2 = 101
N = 1000

position = np.empty([N,2])
position[0,0] = int(L1/2)
position[0,1] = int(L2/2)
direction = [[0,1],[0,-1],[1,0],[-1,0]]

# random walk
for i in range(1,N):
    position1 = [-1,-1]
    
    while min(position1)<0 or position1[0] >L1 or position1[1] >L2:
        move = np.random.randint(len(direction))
        position1 = position[i-1,:] + direction[move]
        
    position[i,:] = position1
plt.xlim(0,101)
plt.ylim(0,101)
plt.title('Brownian motion')
plt.plot(position[:,0], position[:,1], '-o')        
plt.show()
```
