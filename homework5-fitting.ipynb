# Enzyme activity and temperature

Enzymatic activity is known to be affected by by temperature. A study examined the activity rate(in micromoles per second, $ \mu $ mol/s) of the digestive enzyme acidd phosphatase in vitro at varying temperatures(meaured in kelvins, K). The findings are displayed below. (For reference, the range of $ 253 $ to $ 298 K $ corresponds to roughly $ 25 $ to $ 80 degrees Celsius $, or $ 77 $ to $ 176 
 degrees Fahrenheit $.)
(a) Make a scatterplot. (Which is the explanatory variable?)
(b) Describe the form of the relationship. It is not linear. Explan why the form of the relationship makes sense.

Consider the given data, $ Temperature $ in Kelvin, with the corresponding $ Rate $, in $ \mu $ mol/s.

```python
temp = [298, 298, 298, 303, 303, 303, 308, 308, 308, 313, 313, 313, 318, 318, 318, 323, 323, 323, 328, 328, 328, 333, 333, 333, 338, 338, 338, 343, 343, 343, 348, 348, 348, 353, 353, 353]
rate = [.04, .05, .05, .08, .08, .08, .11, .13, .11, .18, .2, .19, .34, .34, .35, .47, .48, .48, .78, .78, .79, .97, .98, 1.03, 1.05, 1.1, 1.06, .96, .98, .99, .72, .73, .75, .54, .58, .62]
```

From what I have read in the description, the activity rate highly depends on the temperature. Therefore, this implies that the temperature must be the independent variable, and the rate is the dependent variable. Let us make a scatterplot to see what type or relationship there is between the two variables.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.scatter(temp, rate)
plt.xlabel('temp')
plt.ylabel('rate')
plt.legend(['rate'], loc = 'upper left')
plt.show()
```

The relationship is definitely not linear. For those who have taken a statistics course, they may be familiar with the Normal Distribution function. Well, this curve looks like it can be modeled by the Normal Distribution function, but it would be weird to do so because I honestly don't know anyone who would use that. The next best thing would probably be to pick a quadratic model, because the plot looks more quadratic than linear. So let us try that. 

```python
coefficients = np.polyfit(temp, rate, deg = 2)
print(coefficients)
def quadModel(t):
    return coefficients[0]*t**2 + coefficients[1]*t + coefficients[2]
```

Let us put this function to the test by plotting another graph using it.

```python
x = np.linspace(298, 353, 36)
y = quadModel(x)
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y'], loc = 'upper left')
plt.show()
```

This doesn't look so bad. While this may not be the best fit, it is significantly better than what we would have gotten if we assumed the model to be linear and used linear regression instead. Let us look at another example.

# Dynamic lift

A study was conducted at Virginia Tech to determine if certain static arm-strength measures have an influence on the "dynamic lift" characteristics of an individual. Twenty-five individuals were subjected to strength tests and then wer asked to perform a weight-lifting test in which weight was dynamically lifted over-head. Let $ x $ represent the Arm Strength, and $ y $ the Dynamic Lift. The data are given here.

```python
x = [17.3, 19.3, 19.5, 19.7, 22.9, 23.1, 26.4, 26.8, 27.6, 28.1, 28.2, 28.7, 29, 29.6, 29.9, 29.9, 30.3, 31.3, 36, 39.5, 40.4, 44.3, 44.6, 50.4, 55.9]
y = [71.7, 48.3, 88.3, 75, 91.7, 100, 73.3, 65, 75, 88.3, 68.3, 96.7, 76.7, 78.3, 60, 71.7, 85, 85, 88.3, 100, 100, 100, 91.7, 100, 71.7]
```

Let us plot the data and see what it looks like.

```python
plt.scatter(x, y)
plt.xlabel('Arm Strength')
plt.ylabel('Dynamic Lift')
plt.legend(['Dynamic Lift'], loc = 'lower right')
plt.show()
```

While the data points appear to be randomly scattered, a statistics student like myself would argue that the model would be best represented by a linear model. Therefore, a linear regression is most appropriate for this type of problem. So let us fit the model. 

```python
constants = np.polyfit(x, y, 1)
print(constants)
def h(t):
    return constants[0]*t + constants[1] # defining the model

# Let us test this model
xi = np.linspace(17.3, 55.9, len(x))
yi = h(xi)
plt.scatter(x, y)
plt.plot(xi, yi, 'r')
plt.xlabel('Arm Strength')
plt.ylabel('Dynamic Lift')
plt.legend(['Regression Line'])
plt.show()
```

Supposed we are asked to estimate the value of the Dynamic Lift when to Arm Strength is $ x = 100 $. What we can do is just plug in that number to the linear function I defined earlier. So let's go ahead and do that.

```python
print('When the Arm Strength is 100, the Dynamic Lift will be', h(100))
```

At an Arm Strength of about $ 100 $, the corresponding Dynamic Lift is roughly $ 121 $.
