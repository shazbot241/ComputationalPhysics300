```python
from collections import OrderedDict
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np

dataset = load_iris()
X = dataset.data
Y = dataset.target
print(X)
print(Y)
```

```python
X1 = np.array(X > X.mean(), dtype='int')
print(X1)
```

```python
Iris_data1 = OrderedDict({'Sepal Length ': X1[:,0],
             'Sepal Width': X1[:,1],
             'Petal Length': X1[:,2],
             'Petal Width': X1[:,3],
             'Label': Y})

df = pd.DataFrame(Iris_data1)
df
```

### If  type '0'

```python
index, SW, PL, PW = 0, 0, 0, 0

while Y[index] == 0:
    if X1[index][1] == 1:
        SW += 1
    if X1[index][2] == 1:
        PL += 1
    if X1[index][3] == 1:
        PW += 1
    index +=1
    
print (SW, PL, PW)
```

### If type '1'

```python
SW, PL, PW = 0, 0, 0

while Y[index] == 1:
    if X1[index][1] == 1:
        SW += 1
    if X1[index][2] == 1:
        PL += 1
    if X1[index][3] == 1:
        PW += 1
    index +=1
    
print (SW, PL, PW)
```
