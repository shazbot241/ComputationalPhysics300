```python
import math
import numpy as np
import matplotlib.pyplot as plt # I may need to use one or all of these libraries later
%matplotlib inline
plt.rcParams['figure.figsize'] = (35,30) # changing figure. I choose 35 by 30. It's just personal preference
plt.rcParams.update({'font.size': 25}) # for me this font size is the sweetspot

if __name__ == "__main__":
    h = 4
    blocks = 500
    spacing = h / blocks
    Mandelbrot = np.empty([blocks, blocks], float) # creating an empty array that we will be filling up as we go on
    
    i = 0
    while i < blocks:
        verticalSpace = spacing*(i - blocks/2)
        for j in range(blocks):
            horizontalSpace = spacing*(j - blocks/2)
            z, n = 0, 0
            
            while True: # keep looping until the expression becomes false and we break out of it
                z = z**2 + complex(horizontalSpace, verticalSpace)
                n = n + 1
                
                if abs(z) >= 2:
                    Mandelbrot[i, j] = n
                    break
                elif n == 180:
                    Mandelbrot[i, j] = 0
                    break
        i = i + 1 # we have to make sure that we increment i or we will be running an infinite loop
        
    plt.imshow(Mandelbrot, aspect = 'auto', cmap = 'hot') # we now want to know what the image looks like
    plt.title('Mandelbrot Set')
    plt.xlabel('X Axis')
    plt.ylabel('Y Axis')
    plt.colorbar()
    plt.show()
```
